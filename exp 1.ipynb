**1. Install Python and understand the difference between scripting and interactive modes in IDLE.**

   **Interactive Mode:**
   Interactive programming means executing code directly in the Python terminal Each line of code runs immediately and displays the output instantly. In this mode, writing long or continuous interlinked programs is not practical. There is also no need to save the code before running it.

   **Script Mode:**
   In script mode, the complete program is written in a file and then executed all at once The file must be saved before execution To run a Python script, we use a terminal command such as **`python <file_name>`**, which instructs the Python interpreter to execute the entire program.
