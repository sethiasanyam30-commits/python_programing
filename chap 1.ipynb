**Varriable - Native Python?**

A variable is a named memory location used for storing values and can be updated modified later on while running a programming language for a certain period of time.

`Naming and Using Variables:`

Variable names in Python may consist of letters digits and underscores They should start with a letter or underscore but not a digit For example **message_1** is valid whereas **1_message** is not.

Variable names cannot contain spaces but underscores may be used to separate words For example **greeting_message** is fine but **greeting message** will result in an error.

It is better to avoid using Python keywords or names of built-in functions such as **print** for naming variables since these words have special meanings in the language.

Variable names must be short and indicative For instance **name** is better than **n** **student_name** is preferable to **s_n** and **name_length** is more comprehensible than **length_of_persons_name**.

Also, be careful with letters such as lowercase **l** and uppercase **O** because they can be confused with the numbers **1** and **0** respectively.

***Note: Python variable names should be written in lowercase Using the capital letters will not result in an error however using lowercase will be better to maintain coherence and readability.***

**Python String Built-in Methods (Full List)**

**1) Case Conversion Methods**

`lower()`

**Definition and Usage**

Return a copy of the string with all the cased characters converted to lowercase. For example:

```python
a="SANYAM"
a.lower()
```

`upper()`

**Definition and Usage**

This method creates and returns a **new string** in which all characters that have case are converted to **uppercase**. Keep in mind that calling `s.upper().isupper()` may still return `False` if the original string contains characters that do not have case, or if the Unicode category of the converted characters is not classified as **uppercase letters (“Lu”)**, but instead falls under another category such as **titlecase letters (“Lt”)**.


```python
"sanyam".upper()
```

`title()`

**Definition and Usage**

This method returns a **title-cased form of the string**, in which each word begins with an **uppercase letter**, while all the other characters in the word are converted to **lowercase**.


```python
"python programming project".title()
```

`capitalize()`

**Definition and Usage**

The `capitalize()` method returns a new string where only the first character is converted to a capital (titlecase) form and all remaining characters are changed to lowercase.

**Updated in Python 3.8:**
The first character is now converted to titlecase instead of uppercase As a result, characters such as digraphs will have only their initial letter capitalized rather than transforming the entire character.


```python
"python project".capitalize()
```

`swapcase()`

**Definition and Usage**

The `swapcase()` method returns a new string in which all uppercase letters are converted to lowercase and all lowercase letters are converted to uppercase.


```python
"pYthOn".swapcase()
```
