# **write a program to print all the digits of a 6 digit number no digit is 0**

```python
num = 123456
while num > 0:
    digit = num % 10
    print(digit)
    num //= 10
```

```python
num = 100001

temp = num

if num < 100000 or num > 999999:
    print("Not a 6-digit number")
else:
    while temp > 0:
        digit = temp % 10
        print(digit)
        temp = temp // 10
```

```python
num = 153
temp = num
sum = 0
while temp > 0:
    digit = temp % 10
    sum = sum + (digit ** 3)
    temp = temp // 10
if sum == num:
    temp = num
    while temp > 0:
        digit = temp % 10
        print(digit)
        temp = temp // 10
```
