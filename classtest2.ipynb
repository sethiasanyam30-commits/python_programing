# q2

```python
class Employee:
    company = "ABC Ltd"

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        print(self.name, self.salary, Employee.company)

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

e1 = Employee("Sanyam", 50000)
e2 = Employee("Riya", 60000)

print("Before:")
e1.show_details()
e2.show_details()

Employee.change_company("XYZ")

print("After:")
e1.show_details()
e2.show_details()
```

# q3

```python
class MathHelper:
    @staticmethod
    def is_even(num):
        """Returns True if num is even, otherwise False."""
        return num % 2 == 0

print(MathHelper.is_even(8)) 
print(MathHelper.is_even(11))
```
