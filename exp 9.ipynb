**Question1. Create a class of student (name, sap id, marks[phy,chem,maths]). Create 3 objects by taking inputs from the user and display details of all students.**

```python

# Create student class
class Student:
    # Constructor to initialize values
    def __init__(self, name, sap_id, phy, chem, maths):
        self.name = name
        self.sap_id = sap_id
        self.phy = phy
        self.chem = chem
        self.maths = maths

    # Function to display student details
    def display(self):
        print("\nStudent Details:")
        print("Name:", self.name)
        print("SAP ID:", self.sap_id)
        print("Physics Marks:", self.phy)
        print("Chemistry Marks:", self.chem)
        print("Maths Marks:", self.maths)


# Create a list to store student objects
students = []

# Taking input for 3 students
for i in range(3):
    print(f"\nEnter details of Student {i+1}")
    
    name = input("Enter name: ")
    sap_id = input("Enter SAP ID: ")
    phy = int(input("Enter Physics marks: "))
    chem = int(input("Enter Chemistry marks: "))
    maths = int(input("Enter Maths marks: "))
    
    # Create object and store in list
    s = Student(name, sap_id, phy, chem, maths)
    students.append(s)

# Display details of all students
print("\n----- All Students Details -----")
for student in students:
    student.display()
```

**Question2: Add constructor in the above class to initialize student details of n students and
implement following methods: Display() student details Find Marks_percentage() of each student Display result() [Note: if marks in each subject >40% than Pass else Fail] Write a Function to find average of the class**

```python

# Function to safely take integer input
def get_marks(subject):
    while True:
        try:
            marks = int(input(f"Enter {subject} marks: "))
            return marks
        except ValueError:
            print("Invalid input! Please enter a number.")

# Student class
class Student:
    
    # Constructor
    def __init__(self, name, sap_id, phy, chem, maths):
        self.name = name
        self.sap_id = sap_id
        self.phy = phy
        self.chem = chem
        self.maths = maths

    # Display student details
    def display(self):
        print("\nStudent Details")
        print("Name:", self.name)
        print("SAP ID:", self.sap_id)
        print("Physics:", self.phy)
        print("Chemistry:", self.chem)
        print("Maths:", self.maths)

    # Calculate percentage
    def marks_percentage(self):
        return (self.phy + self.chem + self.maths) / 3

    # Pass / Fail result
    def result(self):
        if self.phy > 40 and self.chem > 40 and self.maths > 40:
            return "PASS"
        else:
            return "FAIL"


# Function to calculate class average
def class_average(students):
    total = 0
    for s in students:
        total += s.marks_percentage()
    avg = total / len(students)
    print("\nClass Average Percentage:", avg)


# Input number of students
n = int(input("Enter number of students: "))

students = []

# Input details of n students safely
for i in range(n):
    print(f"\nEnter details of Student {i+1}")
    
    name = input("Enter Name: ")
    sap_id = input("Enter SAP ID: ")
    
    phy = get_marks("Physics")
    chem = get_marks("Chemistry")
    maths = get_marks("Maths")
    
    s = Student(name, sap_id, phy, chem, maths)
    students.append(s)

# Display report
print("\n----- Student Report -----")
for s in students:
    s.display()
    print("Percentage:", s.marks_percentage())
    print("Result:", s.result())

# Display class average
class_average(students)
```

**Question3:Create programs to implement different types of inheritances**

```python

#Single inheritance
class Person:
    def show_person(self):
        print("I am a person")
class Student(Person):
    def show_student(self):
        print("I am a Student")

s = Student()
s.show_person()
s.show_student()


#Multilevel inheritance
class Person:
    def show_person(self):
        print("I am a person")
class Student(Person):
    def show_student(self):
        print("I am a student")
class Result(Student):
    def show_result(self):
        print("I have a result")
r = Result()
r.show_person()
r.show_student()
r.show_result()



#Multiple Inheritance
class Father:
    def skills_father(self):
        print("Father: Driving")

class Mother:
    def skills_mother(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    def skills_child(self):
        print("Child: Studying")

c = Child()
c.skills_father()
c.skills_mother()
c.skills_child()




#Hierarchical inheritance
class Person:
    def show(self):
        print("I am a Person")

class Student(Person):
    def study(self):
        print("Student studies")

class Employee(Person):
    def work(self):
        print("Employee works")

s = Student()
e = Employee()

s.show()
s.study()
e.show()
e.work()
```

**Question4. Create a class to implement method Overriding.**

```python

# Parent class
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Child class (Overriding parent method)
class Dog(Animal):
    def sound(self):   # same method name → overriding
        print("Dog barks")

# Another child class
class Cat(Animal):
    def sound(self):
        print("Cat meows")


# Create objects
a = Animal()
d = Dog()
c = Cat()

# Call same method for all objects
a.sound()
d.sound()
c.sound()
```

**Question5. Create a class for operator overloading which adds two Point Objects where Point has x & y values e.g. if P1(x=10,y=20) P2(x=12,y=15) P3=P1+P2 => P3(x=22,y=35)**

```python

# Create Point class
class Point:
    
    # Constructor to initialize x and y values
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading + operator using __add__()
    def __add__(self, other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point(new_x, new_y)

    # Function to display point
    def display(self):
        print(f"Point(x={self.x}, y={self.y})")


# Create two objects
P1 = Point(10, 20)
P2 = Point(12, 15)

# Add two objects using overloaded +
P3 = P1 + P2

# Display result
print("Point 1:")
P1.display()

print("Point 2:")
P2.display()

print("After Addition (P1 + P2):")
P3.display()
```
