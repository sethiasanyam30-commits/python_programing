**Question1: Find Maximum and Minimum from a list (without using max/min)**

```python

def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < maximum:
            minimum = num
    print("maximum =", maximum)
    print("Minimum = ", minimum)

find_max_min([4,9,1,6,2])
```

**Question2: Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.**

```python
def sum_of_cubes(n):
    total = 0

    for i in range(1,n):
        total += i**3
    return total
print("Sum of cubes =", sum_of_cubes(5))
```

**Question3: Write a Python function to print 1 to n using recursion. (Note: Do not use loop)**

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n)
print_numbers(5)
```

**Question4: Write a recursive function to print Fibonacci series upto n terms.**

```python
def fibonacci(n):
    if n <=1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
def print_fibonacci_terms(n):
    for i in range(n):
        print(fibonacci(i), end =" ")
print_fibonacci_terms(7)
```

**Question5: Write a lambda function to find volume of cone.**

```python
volume_cone = lambda r, h: (1/3) * 3.14 * r**2 *h
print("Volume of cone =", volume_cone(3,5))
```
