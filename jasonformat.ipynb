```python
class Convert:
    
    def run(self, s):
        d = {}
        s = s[1:-1]  # remove { }
        
        for p in s.split(","):
            k, v = p.split(":")
            
            k = k.replace('"', '').strip()
            v = v.replace('"', '').strip()
            
            if v.isdigit():
                v = int(v)
            
            d[k] = v
        
        return d
```
