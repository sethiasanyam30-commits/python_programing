**5.Check whether the quadratic equation has real roots or imaginary roots. Display the roots.**

```python
import cmath  

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

d = b**2 - 4*a*c

root1 = (-b + cmath.sqrt(d)) / (2*a)
root2 = (-b - cmath.sqrt(d)) / (2*a)

if d > 0:
    print("The equation has real and distinct roots.")
elif d == 0:
    print("The equation has real and equal roots.")
else:
    print("The equation has imaginary (complex) roots.")


print("Root 1:", root1)
print("Root 2:", root2)
```

**7.Write a program which takes any date as input and display next date of the calendar e.g. I/P: day=20 month=9 year=2005 O/P: day=21 month=9 year 2005**

```python
import datetime

day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

current_date = datetime.date(year, month, day)

next_date = current_date + datetime.timedelta(days=1)

print("Next Date is: day =", next_date.day, "month =", next_date.month, "year =", next_date.year)
```

 * 8.Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage.
CGPA=percentage/10

CGPA range:

0 to 3.4 -> F

3.5 to 5.0->C+

5.1 to 6->B

6.1 to 7-> B+

7.1 to 8-> A

8.1 to 9->A+

9.1 to 10-> O (Outstanding)

```python
marks = []
for i in range(5):
    mark = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)
total_marks = sum(marks)
percentage = (total_marks / 500) * 100
cgpa = percentage / 10
if cgpa >= 0 and cgpa <= 3.4:
    grade = 'F'
    print(grade)
elif cgpa >= 3.5 and cgpa <= 5.0:
    grade = 'C+'
    print(grade)
elif cgpa >= 5.1 and cgpa <= 6.0:
    grade = 'B'
    print(grade)
elif cgpa >= 6.1 and cgpa <= 7.0:
    grade = 'B+'
    print(grade)
elif cgpa >= 7.1 and cgpa <= 8.0:
    grade = 'A'
    print(grade)
elif cgpa >= 8.1 and cgpa <= 9.0:
    grade = 'A+'
    print(grade)
elif cgpa >= 9.1 and cgpa <= 10.0:
    grade = 'O (Outstanding)'
    print(grade)
else:
    grade = 'Invalid CGPA'
    print(grade)
```
